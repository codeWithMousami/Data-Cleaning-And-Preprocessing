# 3Q What is the count of the missing values in each column of the dataframe? Also find the  overall total missing values and convert it in percentage!


```python
missing_count = df.isna().sum()
print('The count of the missing values is :-->\n\n',missing_count)
total_missing = missing_count.sum().sum()
print('-----------------------Print The Total Missing Count-------------------------------------\n\n')
print('The  total count of the missing values is :',total_missing,'\n')
total_rows, total_columns = df.shape
percentage = (total_missing/(total_rows*total_columns))*100
print('-----------------------Print The Percentage Of Missing Count-------------------------------------\n')
print('The percentage of missing values:-',percentage)
```

# Q.6) Calculate the mean and median for the estimated cost and revised cost. What is your observation?


```python
import numpy as np
import pandas as pd

df = pd.read_csv('Building_Permits.csv',low_memory=False)
mean_A = df['Estimated Cost'].mean()
median_A = df['Estimated Cost'].median()

# Calculate the mean and median of column 'B'
mean_B = df['Revised Cost'].mean()
median_B = df['Revised Cost'].median()

print("1-Mean of column 'Estimated Cost':", mean_A)
print("2-Median of column 'Estimated Cost':", median_A)

print("1-Mean of column 'Revised Cost':", mean_B)
print("2-Median of column 'Revised Cost':", median_B)
```

# Q.7) Create a new cleaned dataframe after applying backward/forward fill with two column names only, namely the Revised cost and the Estimated Cost. Transform the dataframe using the following normalization techniques: min-max scaling and z-score.



```python
from sklearn.preprocessing import MinMaxScaler ,StandardScaler
# Perform forward fill for columns 'A' and 'B'
df['Revised Cost'].fillna(method='ffill', inplace=True)
df['Estimated Cost'].fillna(method='ffill', inplace=True)

# Print the updated DataFrame
df = df[['Revised Cost', 'Estimated Cost']]
print ('A New Cleaned Dataframe After Applying Forward Fill\n')
print('---------------------------------------------------------------------------------------\n')
print(df)
print('---------------------------------------------------------------------------------------\n')

scaler = MinMaxScaler()
min_max_normal=scaler.fit_transform(df)
max_min_scaler =pd.DataFrame(min_max_normal, columns=['Revised Cost (Min-Max)','Estimated Cost (Min-Max)'])
print('Min-Max Scaled Data Frame is :-\n',max_min_scaler)
print('---------------------------------------------------------------------------------------\n')

z_scaler= StandardScaler()
z_normal =z_scaler.fit_transform(df)
z_scaled = pd.DataFrame(z_normal,columns=['Revised Cost (z-score)','Estimated Cost (z-score)'])
print('Z-Score Scaled Data Frame is :-\n',z_scaled)
```
